# 3. Update

#### 3.1 Dropping rows

```python
# number of rows with NA value
df.isnull().sum()
```

```python
# drop rows with NA values
df.dropna(inplace=True)
df.isnull().sum()
```

#### 3.2 Dropping collumns

```python
df.drop('Area code', axis=1, inplace=True)
```

```python
df.columns
```

#### 3.3 Creating calculated columns

```python
df.head()
```

```python
df['New column'] = df['Total eve minutes'] + df['Total night minutes']
```

```python
df.head()
```

#### 3.4 Updating entire column

```python
df['New column'] = 99
```

#### 3.5 Updating single value

```python
df.iloc[0,-1] = 777
```

```python
df.iloc[0, df.columns.get_loc('Churn')] = True
```

#### 3.6 Conditional based update with __Apply__

```python
df['Binary Churn'] = df['Churn'].apply(lambda x : 1 if x == True else 0)
```
