```python
# Basic Scatter plot
X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
plt.scatter(X_data, y_data, c='red', s=100, marker='*')
```

```python
years = np.arange(2007,2023 )
base_kg = 70
weights = (np.random.random(16) * 10).round(0).astype(int) + base_kg

plt.plot(years, weights, c='green', lw=1, linestyle='--', )
plt.show()
plt.plot(years, weights, 'g--')
plt.show()
```

```python
x = ['C++', 'C#', 'Python', 'Java', 'Go']
y = [14, 5, 89, 21, 33]

plt.bar(x, y, color='darkred', width=.5, edgecolor='yellow')
```

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages, edgecolor='yellow', bins=20)
plt.show()
plt.hist(ages, edgecolor='yellow', bins=20, cumulative=True)
plt.show()
```

```python
langs = ['C++', 'C#', 'Python', 'Java', 'Go']
votes = [14, 5, 89, 21, 33]
explodes = [0, 0, 0, 0, .2]

plt.pie(votes, labels=langs)
plt.show()
print('\nNo labels:\n')
plt.pie(votes, labels=None)
plt.legend(labels=langs)
plt.show()
print('\nExplode partition:\n/')
plt.pie(votes, labels=langs, explode=explodes)
plt.show()
print('\nShow pct:\n')
plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=.7)
plt.show()
```

```python
heights = np.random.normal(172, 8, 300)

plt.boxplot(heights)
plt.show()

first = np.linspace(0, 10, 25)
second = np.linspace(10, 200, 25)
third = np.linspace(200, 210, 25)
fourth = np.linspace(210, 230, 25)

print('\nIn each section there is the same qty on values:\n')
plt.boxplot(np.concatenate((first, second, third, fourth)))
plt.show()
```

```python
stock_a = np.random.randint(100, 120, 10)
stock_b = np.random.randint(90, 130, 10)
stock_c = np.random.randint(100, 110, 10)

plt.plot(stock_a, label='Google')
plt.plot(stock_b, label='Facebook')
plt.plot(stock_c, label='Tesla')

plt.legend(loc='lower center')

plt.show()
```

### Subplots

```python
X = np.arange(100)

# fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(40, 20))
fig, ax = plt.subplots(2, 2, figsize=(60, 20),)

# ax[r][c]
ax[0][0].plot(X, np.sin(X), c='b')
ax[0][0].set_title('Sine Wave', fontsize=30)

ax[0][1].plot(X, np.cos(X), c='r')
ax[0][1].set_title('Cosine Wave', fontsize=30)

ax[1][0].plot(X, np.random.random(100), c='y')
ax[1][0].set_title('Random', fontsize=30)

ax[1][1].plot(X, np.log(X), c='g')
ax[1][1].set_title('Log Function', fontsize=30)

fig.suptitle('Subplots Demo', fontsize=50)

plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)


ax.set_title('3D Plot')

# Set the axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


plt.show()
```
